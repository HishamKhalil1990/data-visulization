# house price dataset analysis

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
house_df = pd.read_csv("./kc_house_data.csv",error_bad_lines=False)
```

## house prices distribution and spread

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.title("House price distribution")
sns.distplot(house_df.price)
plt.subplot(1,2,2)
plt.title("House price spread")
sns.boxplot(y = house_df.price)
plt.show
```

```python
print(house_df.price.describe())
```

## There is a significant difference between the mean and the median of the price distribution.

## about 75% of the house prices under 645,000, whereas the remaining 25% are between 645,000 and 7,700,000.

## average house price vs. year built

```python
df = pd.DataFrame(house_df.groupby(['yr_built'])['price'].mean().sort_index())
df.plot.bar(figsize=(20,6))
plt.title('year bulit vs Average Price')
plt.show()
```

```python
house_df['yr_built'].mode().item()
```

## the most common built year for houses is 2014

```python
sns.pairplot(house_df[['price','bedrooms','bathrooms','sqft_living','sqft_lot','floors','yr_built']])
plt.show()
```

## linear regression for (aqft_living vs price)

```python
X = house_df["sqft_living"].values.reshape(-1,1)
y = house_df["price"].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 4)
model = LinearRegression()
model.fit(X_train, y_train)
```

## model accurecy

```python
accurcy = model.score(X_test,y_test)
accurcy*100
```

## model vs. scattered data

```python
plt.scatter(X,y,color = "blue")
plt.plot(X,model.predict(X), color = "red")
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.show()
```

## as we can see the house price is related to other dataset in some patteren

## for some patteren, linear regression can be used to create a model can predict the price
