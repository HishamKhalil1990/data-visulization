## car sold quantity data analysis

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
car_sales_df = pd.read_csv("norway_new_car_sales_by_month.csv",error_bad_lines=False)
```

## cars Quantity sold distribution and spread

```python
df = pd.DataFrame(car_sales_df.groupby(['Month'])['Quantity'].mean().sort_index())
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.title("Quantity sold distribution")
sns.distplot(df.Quantity)
plt.subplot(1,2,2)
plt.title("Quantity sold spread")
sns.boxplot(y = df.Quantity)
plt.show()
```

## distribution for car quantity sold is close to normal distribution

```python
print(car_sales_df.Quantity.describe())
```

## the mean and median are close to each others 

## about 75% of the house prices under 12,337, whereas the remaining 25% are between 12,337 and 14,207.

```python
df = pd.DataFrame(car_sales_df.groupby(['Month'])[['Quantity','Used','Avg_CO2','Bensin_Co2','Diesel_Co2','Quantity_Hybrid','Quantity_Electric']].mean().sort_index())
sns.pairplot(df[['Quantity','Used','Avg_CO2','Bensin_Co2','Diesel_Co2','Quantity_Hybrid','Quantity_Electric']])
plt.show()
```

```python
df.plot.bar()
```

## linear regression for (Quantity vs Quantity_Hybrid)

```python
X =df["Quantity"].values.reshape(-1,1)
y = df["Quantity_Hybrid"].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 4)
model = LinearRegression()
model.fit(X_train, y_train)
```

## model accurecy

```python
accurcy = model.score(X_test,y_test)
accurcy*100
```

## model vs. scattered data

```python
plt.scatter(X,y,color = "blue")
plt.plot(X,model.predict(X), color = "red")
plt.xlabel('Quantity')
plt.ylabel('Quantity_Hybrid')
plt.show()
```

## linear regression is unuseful for this case

## the data is scattered, another type of regression can be used to create a model that can predict the quatity to be sold in the future
